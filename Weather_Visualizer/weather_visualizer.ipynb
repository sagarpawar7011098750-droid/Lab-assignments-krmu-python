```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- TASK 1: Data Loading ---
try:
    df = pd.read_csv('weather_data.csv')
    print("Data Loaded Successfully.")
except FileNotFoundError:
    print("File not found. Please ensure 'weather_data.csv' is in the directory.")

# --- TASK 2: Cleaning ---
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Handle missing values 
df = df.dropna() 

# --- TASK 3: Statistics ---
print("Temperature Mean:", np.mean(df['Temperature']))
print("Max Rainfall:", np.max(df['Rainfall']))

# --- TASK 4: Visualization ---
plt.figure(figsize=(10, 6))

plt.plot(df['Date'], df['Temperature'], label='Temp', color='orange')
plt.title('Daily Temperature Trends')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.savefig('temperature_trend.png') 
plt.show()

# --- TASK 5: Grouping ---

df['Month'] = df['Date'].dt.month
monthly_stats = df.groupby('Month')['Rainfall'].sum()
print(monthly_stats)

# --- TASK 6: Export ---

df.to_csv('cleaned_weather_data.csv', index=False)
```
