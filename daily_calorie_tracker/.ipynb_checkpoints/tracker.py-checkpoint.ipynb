```python
# tracker.py
# Daily_Calorie_Tracker_CLI
# Name: Sagar Pawar
# Date: 9 nov 2025
# Project Title: Building a Calorie Tracking

print("Welcome to the Daily Calorie Tracker CLI!")
print("This tool helps you record your meals, calculate total calories, and compare them to your daily limit.\n")

# Task 2: Input & Data Collection
meals = []
calories = []

num_meals = int(input("How many meals did you have today? "))

for i in range(num_meals):
    meal_name = input(f"Enter meal {i+1} name: ")
    meal_calories = float(input(f"Enter calories for {meal_name}: "))
    meals.append(meal_name)
    calories.append(meal_calories)

# Task 3: Calorie Calculations
total_calories = sum(calories)
average_calories = total_calories / num_meals
daily_limit = float(input("\nEnter your daily calorie limit: "))

# Task 4: Exceed Limit Warning System
if total_calories > daily_limit:
    status = "Warning: You have exceeded your daily calorie limit!"
else:
    status = "Good job! You are within your daily calorie limit."

# Task 5: Neatly Formatted Output
print("\nYour Daily Calorie Summary")
print("-" * 40)
print("Meal Name\tCalories")
print("-" * 40)

for i in range(num_meals):
    print(f"{meals[i]}\t\t{calories[i]}")

print("-" * 40)
print(f"Total:\t\t{total_calories}")
print(f"Average:\t{average_calories:.2f}")
print(status)

# Task 6 (Bonus): Save Session Log to File
save = input("\nDo you want to save this report to a file? (yes/no): ").strip().lower()
if save == "yes":
    import datetime
    filename = "calorie_log.txt"
    with open(filename, "a") as f:  # Append mode to keep all sessions
        f.write("\n==============================\n")
        f.write("Daily Calorie Tracker Report\n")
        f.write(f"Date: {datetime.datetime.now()}\n\n")
        f.write("Meal Name\tCalories\n")
        f.write("-" * 40 + "\n")
        for i in range(num_meals):
            f.write(f"{meals[i]}\t\t{calories[i]}\n")
        f.write("-" * 40 + "\n")
        f.write(f"Total:\t\t{total_calories}\n")
        f.write(f"Average:\t{average_calories:.2f}\n")
        f.write(status + "\n")
        f.write("==============================\n")
    print(f"\nReport saved successfully and added to '{filename}'.")
else:
    print("Report not saved. Session ended.")
```
