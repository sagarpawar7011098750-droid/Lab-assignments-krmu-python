```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

# --- TASK 3: OOP DESIGN  ---

class MeterReading:
    def __init__(self, timestamp, kwh):
        self.timestamp = timestamp
        self.kwh = kwh

class Building:
    def __init__(self, name):
        self.name = name
        self.readings = [] # List to store MeterReading objects

    def add_reading(self, reading_obj):
        self.readings.append(reading_obj)

    def get_dataframe(self):
        """Helper to convert readings list to Pandas DataFrame for aggregation"""
        data = [{'timestamp': r.timestamp, 'kwh': r.kwh} for r in self.readings]
        df = pd.DataFrame(data)
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df.set_index('timestamp', inplace=True)
        return df

# --- TASK 1: DATA INGESTION [cite: 5, 6, 8] ---

def load_data(data_folder):
    buildings = [] # List of Building objects
    
    # search for csv files
    files = glob.glob(os.path.join(data_folder, "*.csv"))
    
    if not files:
        print("No CSV files found!")
        return []

    for file_path in files:
        try:
            # Extract building name from filename (e.g., 'Library.csv' -> 'Library')
            building_name = os.path.basename(file_path).replace('.csv', '')
            
            # Create Building Object
            current_building = Building(building_name)
            
            # Read CSV using Pandas [cite: 7]
            df_raw = pd.read_csv(file_path, on_bad_lines='skip') # Handle corrupt lines [cite: 8]
            
            # Populate Building object
            for _, row in df_raw.iterrows():
                # Assuming CSV has 'Timestamp' and 'Usage' columns
                reading = MeterReading(row['Timestamp'], row['Usage'])
                current_building.add_reading(reading)
            
            buildings.append(current_building)
            print(f"Successfully loaded {building_name}")

        except Exception as e:
            print(f"Error loading {file_path}: {e}")
            
    return buildings



def analyze_and_plot(buildings):
    master_df = pd.DataFrame()

    # --- 1. Flatten Data for Analysis ---
    # Convert our list of Building objects into one big DataFrame
    for b in buildings:
        df = b.get_dataframe()
        if not df.empty:
            df['Building'] = b.name
            master_df = pd.concat([master_df, df])
    
    if master_df.empty:
        print("No data available to plot.")
        return master_df

    # --- 2. Visualization Setup (Task 4) ---
    # Create a figure with 3 subplots (rows)
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 18))
    plt.subplots_adjust(hspace=0.4) # Add space between charts

    # --- CHART 1: Trend Line (Daily Consumption) ---
    # Resample to Daily Sum per Building
    daily_data = master_df.groupby('Building').resample('D')['kwh'].sum().reset_index()
    
    for name in daily_data['Building'].unique():
        subset = daily_data[daily_data['Building'] == name]
        ax1.plot(subset['timestamp'], subset['kwh'], label=name, marker='o', markersize=3)
    
    ax1.set_title('Daily Energy Consumption Trends', fontsize=14)
    ax1.set_ylabel('Total kWh (Daily)')
    ax1.set_xlabel('Date')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # --- CHART 2: Bar Chart (Average Weekly Usage) ---
    # Resample to Weekly Mean per Building
    weekly_data = master_df.groupby('Building').resample('W')['kwh'].mean().reset_index()
    avg_weekly = weekly_data.groupby('Building')['kwh'].mean() # Average across all weeks
    
    avg_weekly.plot(kind='bar', ax=ax2, color=['#4c72b0', '#55a868', '#c44e52'])
    
    ax2.set_title('Average Weekly Electricity Usage by Building', fontsize=14)
    ax2.set_ylabel('Average kWh')
    ax2.set_xlabel('Building')
    ax2.tick_params(axis='x', rotation=0)
    ax2.grid(axis='y', alpha=0.3)

    # --- CHART 3: Scatter Plot (Peak-Hour Consumption) ---
    # Extract hour from timestamp
    master_df['Hour'] = master_df.index.hour
    
    # Define "Peak" as generally high consumption (e.g., > 75th percentile) or just plot all
    # Plotting 'Hour' vs 'Usage' to show when spikes happen
    
    for name in master_df['Building'].unique():
        subset = master_df[master_df['Building'] == name]
        ax3.scatter(subset.index, subset['kwh'], label=name, alpha=0.6, s=10)

    ax3.set_title('Consumption Distribution Over Time (Spotting Peaks)', fontsize=14)
    ax3.set_ylabel('kWh Usage')
    ax3.set_xlabel('Date/Time')
    ax3.legend()
    ax3.grid(True, alpha=0.3)

    # Save and Show
    plt.savefig('dashboard.png')
    print("Dashboard saved as 'dashboard.png'.")
    plt.show() # This will make it show up in your Jupyter Notebook
    
    return master_df
# --- MAIN EXECUTION ---
if __name__ == "__main__":
    # Ensure you have a folder named 'data' with csv files in it
    all_buildings = load_data('data')
    
    if all_buildings:
        full_data = analyze_and_plot(all_buildings)
        
        # TASK 5: EXPORT [cite: 19]
        full_data.to_csv("cleaned_energy_data.csv")
        # Write summary.txt logic here...
        
def generate_summary_report(df):
    if df.empty:
        return

    # Calculate required stats [cite: 81, 82, 83]
    total_consumption = df['kwh'].sum()
    highest_building = df.groupby('Building')['kwh'].sum().idxmax()
    peak_load_time = df['kwh'].idxmax()
    peak_load_value = df['kwh'].max()
    
    report_content = f"""
    CAMPUS ENERGY EXECUTIVE SUMMARY
    ===============================
    
    1. Total Campus Consumption: {total_consumption:.2f} kWh
    2. Highest Consuming Building: {highest_building}
    3. Peak Load Event:
       - Time: {peak_load_time}
       - Usage: {peak_load_value:.2f} kWh
       
    Analysis generated automatically.
    """
    
    # Save to file
    with open('summary.txt', 'w') as f:
        f.write(report_content)
    
    print("Summary report generated: summary.txt")
    print("-" * 30)
    print(report_content)

# CALL THIS IN YOUR MAIN BLOCK:
# generate_summary_report(full_data)
        
```
